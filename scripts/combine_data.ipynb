```python
tract_a = gpd.read_file("../data/alameda_epc.geojson")
tract_sl = gpd.read_file("../data/san_leandro_epc.geojson")
tract_cv = gpd.read_file("../data/castro_valley_epc.geojson")
combined_boundary = pd.concat([tract_a, tract_sl, tract_cv], ignore_index=True)
tract_a = tract_a.to_crs(3857)
tract_sl = tract_sl.to_crs(3857)
tract_cv = tract_cv.to_crs(3857)
combined_boundary = combined_boundary.to_crs(3857)

fig = plt.figure(figsize=[12,8])
ax = fig.add_axes([0, 0, 1, 1])
combined_boundary.plot(ax=ax)

cx.add_basemap(ax, source=cx.providers.CartoDB.Positron)
ax.set_axis_off()
```

```python
# collect all the proccessed data
bus_a = gpd.read_file("../data/bus_stops_alameda.geojson")
bus_sl = gpd.read_file("../data/bus_stops_san_leandro.geojson")
bus_cv = gpd.read_file("../data/bus_stops_castro_valley.geojson")

bus_a = bus_a.to_crs(3857)
bus_sl = bus_sl.to_crs(3857)
bus_cv = bus_cv.to_crs(3857)

t2n_a = gpd.read_file("../data/t2n_alameda.geojson")
t2n_sl = gpd.read_file("../data/t2n_san_leandro.geojson")
t2n_cv = gpd.read_file("../data/t2n_castro_valley.geojson")

t2n_a = t2n_a.to_crs(3857)
t2n_sl = t2n_sl.to_crs(3857)
t2n_cv = t2n_cv.to_crs(3857)

schools_a = gpd.read_file("../data/alameda_schools.geojson")
schools_sl = gpd.read_file("../data/san_leandro_schools.geojson")
schools_cv = gpd.read_file("../data/castro_valley_schools.geojson")

schools_a = schools_a.to_crs(3857)
schools_sl = schools_sl.to_crs(3857)
schools_cv = schools_cv.to_crs(3857)
```

```python
# plot it all together: Alameda
fig = plt.figure(figsize=[12,8])
ax = fig.add_axes([0, 0, 1, 1])

bus_a.plot(ax=ax, color='blue', label='bus stops')
schools_a.plot(ax=ax, color='red', label = 'schools')
t2n_a.plot(ax=ax, color='yellow', label='pedestrian paths')

cx.add_basemap(ax, source=cx.providers.CartoDB.Positron)
ax.set_axis_off()
plt.legend()
```

```python
# plot it all together: San Leandro
fig = plt.figure(figsize=[12,8])
ax = fig.add_axes([0, 0, 1, 1])

bus_sl.plot(ax=ax, color='blue', label='bus stops')
schools_sl.plot(ax=ax, color='red', label = 'schools')
t2n_sl.plot(ax=ax, color='yellow', label='pedestrian paths')

cx.add_basemap(ax, source=cx.providers.CartoDB.Positron)
ax.set_axis_off()
plt.legend()
```

```python
# plot it all together: Castro Valley
fig = plt.figure(figsize=[12,8])
ax = fig.add_axes([0, 0, 1, 1])

bus_cv.plot(ax=ax, color='blue', label='bus stops')
schools_cv.plot(ax=ax, color='red', label = 'schools')
t2n_cv.plot(ax=ax, color='yellow', label='pedestrian paths')

cx.add_basemap(ax, source=cx.providers.CartoDB.Positron)
ax.set_axis_off()
plt.legend()
```
